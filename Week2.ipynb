```
#1. -Tampilkan image dalam channel Red-Green !
img_rg = img.copy()
img_rg[:,:,2] = 0
plt.imshow(img_rg)
```

```
# -Tampilkan image dalam channel Green-Blue
img_gb = img
img_gb[:,:,0] = 0
plt.imshow(img_gb)
```

```
#2. Tampilkan image baris ke 10-100, kolom 10-100!
img_k = img[10:100,10:100]
plt.imshow(img_k)
```

```
#3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
plt.imshow(img[5:30])
```

```
#4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
imageWithRectangle = cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,10))
cv.rectangle(imageWithRectangle, pt1=(1000,800), pt2=(400,95), color=(0,255, 255), thickness = 20)
cv.rectangle(imageWithRectangle, pt1=(1400,1200), pt2=(2200,500), color=(0,255,0), thickness = 40)
cv.rectangle(imageWithRectangle, pt1=(2000,1400), pt2=(700,400), color=(255,255,255), thickness = 20)
cv.rectangle(imageWithRectangle, pt1=(200,1500), pt2=(400,900), color=(255,0,0), thickness = 25)
cv.rectangle(imageWithRectangle, pt1=(500,150), pt2=(1200,750), color=(0,0,255),thickness=17)
plt.imshow(imageWithRectangle)
```

```
#5. Tampilkan image dengan posisi terbalik!
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

```
#6. Buat rectangle dan circle pada bagian wajah dari image kucingku!
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(2300,400), pt2=(1300,1400), color=(255, 0, 30), thickness = 40)
cv.circle(img, center=(1900, 1000), radius=300, color=(180, 105, 255), thickness = 30)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```
#7. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari citra kitten01.jpg
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(2300,400), pt2=(1300,1400), color=(255, 0, 30), thickness = 40)
cv.circle(img, center=(1900, 1000), radius=300, color=(180, 105, 255), thickness = 30)
cv.rectangle(img,pt1=(200, 1500),pt2=(500, 1800),color=(0, 0, 255),thickness = 40)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```
